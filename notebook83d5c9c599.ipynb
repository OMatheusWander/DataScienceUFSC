## Creio que era para ter feito uma tabela dinâmica, ajustado os dados e nisso gastei horas tentando rodar um código não limpo

#tinha colunas de anos que era para ter sido transformadas em linhas

```python
import pandas as pd

# Suponha que você tenha um DataFrame df com as colunas 'Ano' e 'Valor'
data = {'Ano': [2019, 2020, 2021],
        'Valor': [100, 200, 150]}

df = pd.DataFrame(data)

# Usando a função 'explode' para transformar os valores em linhas
df = df.assign(Valor=df['Valor'].str.split(',')).explode('Valor')

# Isso irá criar uma nova linha para cada valor separado por vírgula na coluna 'Valor'

# Replicando as informações de 'Ano' para cada linha
df = df.reset_index(drop=True)

# O resultado final terá cada valor em uma linha separada com o ano correspondente
print(df)
```
