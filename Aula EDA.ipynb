```python
import pandas as pd
import numpy as np
filename = "/kaggle/input/noshowappointments/KaggleV2-May-2016.csv"
df = pd.read_csv(filename)
df.head()


df['Neighbourhood'] 
```

```python
com_bolsa_df = df[df['Scholarship'] == 1]

com_bolsa_df
```

```python
bairroebolsa = df.groupby(['Neighbourhood'])['Scholarship'].agg([np.sum])
bairroebolsa
```

```python
bairroebolsa.columns
```

```python
bairroebolsa.groupby(['Neighbourhood'])['sum'].agg([np.sum]).sort_values(by='sum',ascending=False)#.index[0]
```

## 5 -Qual é o bairro que recebeu maior número de pacientes?

```python
bairroseshow = df[df['No-show'] == "No"]
bairroseshow.columns
```

```python
bairroseshow.groupby(['Neighbourhood'])['PatientId'].agg([np.count_nonzero]).sort_values(by=['count_nonzero'], ascending=False).index[0]
```

## 6- Qual gênero tem mais bolsas?

```python
com_bolsa_df.groupby(['Gender'])['PatientId'].agg([np.count_nonzero]).sort_values(by=['count_nonzero'], ascending=False).index[0]
```

## 7- Qual é a média da idades dos pacientes: que são Homens ou Mulheres?


```python
df_homens = df[df['Gender'] == "M"]
df_homens['Age'].mean()
```

```python
df_mulheres = df[df['Gender'] == "F"]
df_mulheres['Age'].mean()
```
