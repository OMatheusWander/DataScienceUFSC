```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

gfp = "/kaggle/input/military-data/GLOBAL_FIREPOWER.CSV"

df = pd.read_csv(gfp)
df.head()
```

# 1. Qual o país tem a maior relação entre população e pessoal nas forças armadas (total military personal) (2 pontos)




```python
df['Total Population and T. M. Personnel'] = df['Total Military Personnel'] / df['Total Population']
```

```python
df.head()
```

```python
valor_maximo = df['Total Population and T. M. Personnel'].max()
linha_maxima = df[df['Total Population and T. M. Personnel'] == valor_maximo]
linha_maxima
```

# 2. Qual o país que tem a menor relação entre Tanks de combat e Helicópteros? (2 pontos)

```python
df.columns
```

```python
df['Combat tanks/ attack Helicopters'] = df['Combat Tanks']/df['Attack Helicopters']
valor_minimo = df['Combat tanks/ attack Helicopters'].min()

linha_minima = df[df['Combat tanks/ attack Helicopters'] == valor_minimo]
linha_minima
```

# 3. Qual o país que tem a maior relação entre submarinos e orçamento militar (defense budged)? (2 pontos)
