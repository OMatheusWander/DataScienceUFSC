```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

gfp = "/kaggle/input/military-data/GLOBAL_FIREPOWER.CSV"

df = pd.read_csv(gfp)
df.head()
```

# Lista Extra de Data Science

## 1. Qual o país tem a maior relação entre população e pessoal nas forças armadas (total military personal)



```python
df['Population/Military Personnel Ratio'] = df['Total Military Personnel'] / df['Total Population']
```

```python
valor_maximo = df['Population/Military Personnel Ratio'].max()
linha_maxima = df[df['Population/Military Personnel Ratio'] == valor_maximo]
Resp1 = linha_maxima[['Country','Population/Military Personnel Ratio']]
Resp1
```

## 2. Qual o país que tem a menor relação entre Tanks de combat e Helicópteros? 


```python
df.columns
```

```python
df['Combat tanks/attack Helicopters'] = df['Combat Tanks']/df['Attack Helicopters']
valor_minimo = df['Combat tanks/attack Helicopters'].min()

linha_minima = df[df['Combat tanks/attack Helicopters'] == valor_minimo]
linha_minima

Resposta = linha_minima[['Country', 'Combat tanks/attack Helicopters']]

Resposta
```

## 3. Qual o país que tem a maior relação entre submarinos e orçamento militar (defense budged)? 



```python
df.columns 
```

```python
df['Greater relationship between submarines and military budget'] = df['Submarines']/df['Defense Budget']

valor_maximo2 = df['Greater relationship between submarines and military budget'].max()

linha_maxima2 = df[df['Greater relationship between submarines and military budget'] == valor_maximo2]
(linha_maxima2).iloc[0:25,0:1]

```

# 4. Qual a relação entre pessoal ativo (active personal) e pessoal na reserva (reserve personal)?(2 pontos)


```python
df['Relationship active personnel / reserve personnel'] = df['Active Personnel']/df['Reserve Personnel']*100
df1 = df.sort_values(by='Relationship active personnel / reserve personnel', ascending=False)
print(df1[['Country', 'Relationship active personnel / reserve personnel']])
```

## 5.  Quais os três países com maior orçamento militar (defense budged) que tenham os quatro tipos de navios da marinha: submarinos, destroyers, corvets e fragatas?

```python
paises_com_navios = df[
    (df['Submarines'] > 0) &
    (df['Destroyers'] > 0) &
    (df['Corvettes'] > 0) &
    (df['Frigates'] > 0)
]

paises_com_navios = paises_com_navios.sort_values(by='Defense Budget', ascending=False)

top_tres_paises = paises_com_navios.head(3)

print(top_tres_paises[['Country', 'Defense Budget']])
```
